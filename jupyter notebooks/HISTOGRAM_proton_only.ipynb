```python
'''
This script plots histograms of objective function values fitting over 7000 events classified using CNN models.
'''
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

## Differential Evolution Cleaned Events

```python
plt.rcParams["axes.labelsize"] = 16
plt.rcParams["axes.titlesize"] = 16
plt.rcParams["xtick.labelsize"] = 16
plt.rcParams["ytick.labelsize"] = 16
plt.rcParams.update({'font.size': 14})
DE = np.loadtxt("/home/chen/DE_run_0130.txt")
DE_time = np.loadtxt("/home/chen/DE_run_0130_time.txt")
DE_avg = sum(DE)/len(DE)
DE_avgtime = sum(DE_time)/len(DE_time)
print(DE_avg)
print(DE_avgtime)
print(len(DE))
plt.hist(DE, label='Cleaned Tracks', bins=50, color='green',range=(0,140)) 
plt.legend(loc=1)
plt.title('Differential Evolution Fitting Results')
plt.xlabel('Goodness of Fit')
plt.ylabel('Number of Events')
plt.show()
```
